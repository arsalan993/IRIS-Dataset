# IRIS is default detaset in R

```r
class(iris)
```

## Scatter Plot

```r
pairs(iris[1:4], main = "Scatter Plot Iris Data", pch = 21, bg = c("red", "green3", "blue")[unclass(iris$Species)])
```

## Box Plot

```r
data(iris)
# Create separate boxplots for each attribute
par(mfrow=c(2,2))
for(i in 1:4) {
boxplot(iris[,i], main=names(iris)[i])
}
```

## Histogram

```r
# load the data
data(iris)
# create histograms for each attribute
par(mfrow=c(2,2))
for(i in 1:4) {
hist(iris[,i], main=names(iris)[i])
}
```

## Density Plots

```r
library(lattice)
# load dataset
data(iris)
# create a panel of simpler density plots by attribute
par(mfrow=c(2,2))
for(i in 1:4) {
	plot(density(iris[,i]), main=names(iris)[i])
}
```

## Correlation Plot

```r
library(corrplot)
# load the data
data(iris)
# calculate correlations
correlations <- cor(iris[,1:4])
# create correlation plot
corrplot(correlations, method="circle")
```
