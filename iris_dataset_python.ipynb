```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
pylab.rcParams['figure.figsize'] = 10, 8
```

# Dataset

Download dataset from [IRIS dataset](http://archive.ics.uci.edu/ml/machine-learning-databases/iris/). 

Number of Instances: 150 (50 in each of three classes)

Number of Attributes: 4 numeric, predictive attributes and the class

Attribute Information:
1. sepal length in cm
2. sepal width in cm
3. petal length in cm
4. petal width in cm
5. class: 
  * Iris Setosa
  * Iris Versicolour
  * Iris Virginica

```python
iris = pd.read_csv('iris.data', names = ['sepal_length', 'sepal_width',\
        'petal_length', 'petal_width', 'label'], skipfooter = 1, engine = 'python')
```

# Exploratory Data Analysis

### Distribution of each class in dataset[Pie Chart]

```python
iris.groupby(by = 'label').size().plot(kind = 'pie',figsize = (4, 4))
```

### Distribution of each feature variable[Density Plot]

```python
iris.plot(kind = 'kde', figsize = (7, 4))
```

### Aggregrates and summary Statistics:

```python
iris.groupby('label').aggregate([numpy.mean, np.median, np.std])
```

iris.groupby('label').describe() displays all aggregration summary.

### Barplots of aggregration

```python
groups = iris.groupby(by = 'label')
means, sds = groups.mean(), groups.std()
means.plot(yerr = sds, kind = 'bar', figsize = (10,7), table = True)
```

### Boxplots

```python
iris.boxplot(by = "label", figsize = (10, 6))
```

### Feature Plot:

```python
col_map = {'Iris-setosa': 'olive', 'Iris-versicolor': 'navy', 'Iris-virginica': 'magenta'}
pd.plotting.scatter_matrix(iris.loc[:, 'sepal_length':'petal_width']
, diagonal = 'kde', color = [col_map[lb] for lb in iris.label], s = 75)

#print(iris.loc[:, 'sepal_length':'petal_width']) will create data frame with all labels included
```
